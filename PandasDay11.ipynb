```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

X = np.random.rand(100,2)
kmeans = KMeans(n_clusters=3)

kmeans.fit(X)

labels =kmeans.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)

plt.title('actual data points')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.legend()
plt.show()
```

```python
ages = np.random.randint(20,70,size =200)
income = np.random.randint(20000,70000,size =200)

data = np.column_stack((ages,income))
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)
labels = kmeans.predict(data)

plt.scatter(data[:,0],data[:,1], c= labels)

plt.title('ages vs income')
plt.xlabel('age')
plt.ylabel('income')
plt.show()
```

```python
k_values = range(1,11)
intersia_values =[]
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    intersia_values.append(kmeans.inertia_)

plt.scatter(k_values,intersia_values)
plt.xlabel('K')
plt.ylabel('inertia')
```

```python
optimal_k = np.argmin(np.diff(intersia_values))+1
```
